```python
import numpy as np
np.random.seed(1)

import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.random.randint(0, 50, 1000)

# 這次讓 y 與 x 正相關，再增加一些雜訊
y = -x - np.random.normal(0, 50, 1000)

# 再次用 numpy 裡的函數來計算相關係數
np.corrcoef(x, y)
```

```python
plt.scatter(x, y)
```

```python
x = np.random.randint(0, 50, 1000)
y = np.random.randint(0, 50, 1000)

# 呼叫 numpy 裡的相關矩陣函數 (corrcoef)
np.corrcoef(x, y)
```

```python
plt.scatter(x, y)
```
